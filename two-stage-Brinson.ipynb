```python
Rp1=sum(hld_data1['LogReturn']*hld_data1['Weight'])
Rp2=sum(hld_data2['LogReturn']*hld_data2['Weight'])
Rb1=sum(ben_data1['LogReturn']*ben_data1['Weight'])
Rb2=sum(ben_data2['LogReturn']*ben_data2['Weight'])
Rp=(1+Rp1)*(1+Rp2)-1
Rb=(1+Rb1)*(1+Rb2)-1
```

```python
k1=(np.log(1+Rp1)-np.log(1+Rb1))/(Rp1-Rb1)
k2=(np.log(1+Rp2)-np.log(1+Rb2))/(Rp2-Rb2)
K=(np.log(1+Rp)-np.log(1+Rb))/(Rp-Rb)
```

```python
hld_data1['Rtn_Prod']=hld_data1['LogReturn']*hld_data1['Weight']
hld_data2['Rtn_Prod']=hld_data2['LogReturn']*hld_data2['Weight']
ben_data1['Rtn_Prod']=ben_data1['LogReturn']*ben_data1['Weight']
ben_data2['Rtn_Prod']=ben_data2['LogReturn']*ben_data2['Weight']
hld_data1=hld_data1.rename(columns={'FstIndNameCSRS':'Industry'})
hld_data2=hld_data2.rename(columns={'FstIndNameCSRS':'Industry'})
```

```python
ben1=ben_data1[['Industry','Rtn_Prod','Weight']].groupby(['Industry']).sum().reset_index()
ben2=ben_data2[['Industry','Rtn_Prod','Weight']].groupby(['Industry']).sum().reset_index()
hld1=hld_data1[['Industry','Rtn_Prod','Weight']].groupby(['Industry']).sum().reset_index()
hld2=hld_data2[['Industry','Rtn_Prod','Weight']].groupby(['Industry']).sum().reset_index()
```

```python
hld1['Rtn_hld']=hld1['Rtn_Prod']/hld1['Weight']
hld2['Rtn_hld']=hld2['Rtn_Prod']/hld2['Weight']
ben1['Rtn_ben']=ben1['Rtn_Prod']/ben1['Weight']
ben2['Rtn_ben']=ben2['Rtn_Prod']/ben2['Weight']
```

```python
del hld1['Rtn_Prod']
del hld2['Rtn_Prod']
del ben1['Rtn_Prod']
del ben2['Rtn_Prod']
```

```python
ben1
```

```python
ben2
```

```python
data_p1=pd.merge(hld1,ben1,on='Industry',how='outer').rename(columns={'Weight_x':'Weight_hld','Weight_y':'Weight_ben','Rtn_hld_x':'Rtn_hld','Rtn_hld_y':'Rtn_ben'})
data_p2=pd.merge(hld2,ben2,on='Industry',how='outer').rename(columns={'Weight_x':'Weight_hld','Weight_y':'Weight_ben','Rtn_hld_x':'Rtn_hld','Rtn_hld_y':'Rtn_ben'}) 
```

```python
data_p1
```

```python
data_p2
```

```python
data_all=[data_p1,data_p2]
k_list=[k1,k2]
result=pd.DataFrame()
for i in range(3):
    if i<2:
        
        data_all[i]=data_all[i].fillna(0)
        Q=pd.DataFrame()
        Q['Q1'] = data_all[i]['Weight_ben'] * data_all[i]['Rtn_ben']
        Q['Q2'] = data_all[i]['Weight_hld'] * data_all[i]['Rtn_ben']
        Q['Q3'] = data_all[i]['Weight_ben'] * data_all[i]['Rtn_hld']
        Q['Q4'] = data_all[i]['Weight_hld'] * data_all[i]['Rtn_hld']
        Q['AR'] = Q['Q2'] - Q['Q1']
        Q['SR'] = Q['Q3'] - Q['Q1']
        Q['IR'] = Q['Q4'] - Q['Q3'] - Q['Q2'] + Q['Q1']
        Q['TR'] = Q['Q4'] - Q['Q1']
        a = pd.DataFrame({'AR': Q['AR'].sum(),
                      'SR': Q['SR'].sum(),
                      'IR': Q['IR'].sum(),
                      'TR': Q['TR'].sum(),
                         'Period':'P'+str(i)},index=[i])
        result=result.append(a)
    else:
        a = pd.DataFrame({'AR': sum(k_list*result['AR'])/K,
                      'SR': sum(k_list*result['SR'])/K,
                      'IR': sum(k_list*result['IR'])/K,
                      'TR': sum(k_list*result['TR'])/K,
                         'Period':'All'},index=[i])
        result=result.append(a)
    result['FundCode']='270005'
    result=result[['FundCode','AR','SR','IR','TR','Period']]
```
